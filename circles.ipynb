```python
#Autocorrelation function to estimate the period of random data with homogeneous distribution
def corr(arg):
  fftarg = fft(arg, n=(numline*2-1))
  ret = ifft(fftarg * np.conjugate(fftarg))
  return ret
```

```python
#This box allows us to visualize some periodicity of the signal between circles
#And note that in fact, the maximum value of autocorrelation is at the origin
from scipy.fftpack import fft, fftfreq
x = log10(corr(x_obs).real)
y = log10(corr(y_obs).real)
plt.plot(x,marker='o')
plt.plot(y,marker='o')
xlim([-5,450])
ylim([-3,5])
print x
```

```python
#Here one can see the symmetry of the autocorrelation function
x = log10(corr(x_obs).real)
y = log10(corr(y_obs).real)
plt.plot(x,marker='o')
plt.plot(y,marker='o')
xlim([-5,2*numline])
ylim([-3,5])
```
